### Ejercicio 5
Crea un gráfico de barras con barras horizontales utilizando los siguientes datos:
- Categoría A: 7
- Categoría B: 12
- Categoría C: 4
- Categoría D: 2
- Categoría E: 9

El gráfico debe mostrar las frecuencias al final de la barra con una ligera separación de la barra (0.1). Asegúrate de que los números no se superpongan con los bordes.

```python
categories = ['A', 'B', 'C', 'D', 'E']
frequencies = [7, 12, 4, 2, 9]

# Su código va aquí.
bar_width = 0.4  # Ancho original de las barras
bar_padding = 0.1  # Separación deseada entre las barras

# Ajustar las posiciones de las barras
positions = range(len(categories))
adjusted_positions = [pos + bar_padding * i for i, pos in enumerate(positions)]

# Crear la figura y los ejes
fig, ax = plt.subplots()

# Dibujar las barras
ax.bar(adjusted_positions, frequencies, width=bar_width)

# Configuración de ejes y etiquetas
ax.set_xticks([pos + bar_width/2 for pos in adjusted_positions])
ax.set_xticklabels(categories)
ax.set_xlabel('Categorias')
ax.set_ylabel('Frecuencias')
ax.set_title('Grafico de frecuencias')

plt.show()
```

Ahora vamos a jugar con gráficos más avanzados. En particular, vamos a usar gráficos paramétricos, gráficos implícitos, gráficos en 3D, curvas paramétricas en el espacio y superficies paramétricas. Algo así como las cosas que necesitabas cuando tomabas un curso de cálculo, solo que llegaron unos años tarde. Así que primero importemos estas herramientas.

```python
from sympy.plotting import (plot_parametric,plot_implicit,
                            plot3d,plot3d_parametric_line,
                            plot3d_parametric_surface)
```

Ahora, crea el símbolo $t$ y grafica la siguiente curva paramétrica (llamada "Lissajous", deben incluir esto en el título):

$$r(t) = (\sin{(2t)}, \cos{(3t)}), \quad t \in [0, 2\pi)$$

```python
t = Symbol('t')

# Su código va aquí.

# Definir las ecuaciones paramétricas
x = sp.sin(2*t)
y = sp.cos(3*t)

plot_parametric(x, y, (t, 0, 2*sp.pi), title="Lissajous")
```

### Ejercicio 6
Ahora probaremos algunos gráficos paramétricos, primero graficaremos una curva en 3D:
    $$x(t) = \cos{(t)}$$
    $$y(t) = \sin{(t)}$$
    $$z(t) = at, \quad a = 0.1$$
Con $t \in [0, 4\pi]$.

```python
t = Symbol('t')

a = 0.1
x = cos(t)
y = sin(t)
z = a*t

# Su código va aquí.
plot3d_parametric_line(x, y, z, (t, 0, 4*pi))



```

Ahora grafiquemos una superficie paramétrica, definiremos los símbolos $u$ y $v$ para la siguiente superficie:
$$x(u, v) = (1 + a\cos{(u)})\cos{(v)}$$
$$y(u, v) = (1 + a\cos{(u)})\sin{(v)}$$
$$z(u, v) = a\sin{(u)}, \quad a = 0.3$$
Con $u, v \in [0, 2\pi)$.

```python
u,v=symbols('u v')

# Su código va aquí.

a = 0.3
x = (1 + a*cos(u))*cos(v)
y = (1 + a*cos(u))*sin(v)
z = a*sin(u)

plot3d_parametric_surface(x, y, z, (u, 0, 2*pi), (v, 0, 2*pi))
```

## Ejercicio 3
Ahora haremos el mismo gráfico anterior, pero con color. Para ello, usaremos la función `map`, que toma una columna y asigna sus valores a otro de acuerdo a un diccionario. La sintaxis es como sigue:
`colores = datos['columna'].map({'valor_1': 'red', 'valor_2': 'blue'})`
Notar que puede haber más valores y colores, según sea necesario. Aplique esto sobre los datos de `iris` y utilice la columna `species` para definir sus valores. Tendrá que explorar cuales son los posibles valores de dicha columna y asignarles colores. Una vez que tenga los colores, solo debe pasar la lista de colores con el parámetro `c` a la función `plot`

```python
# Su código aquí.

# Saber todos los valores posibles de la columna species.
valores_unicos = iris['species'].unique()
print(valores_unicos)

# Mapear la nueva columna 'color' la cual almacena los colores según las especies.
iris['color'] = iris['species'].map({valores_unicos[0]: 'red', valores_unicos[1]: 'blue', valores_unicos[2]: 'green'})

iris.plot(kind='scatter', x='petal_length', y='petal_width', c='color')

plt.title('Gráfico de Dispersión con Colores por Especie')
plt.xlabel('Longitud del Pétalo')
plt.ylabel('Ancho del Pétalo')

plt.show()

```
